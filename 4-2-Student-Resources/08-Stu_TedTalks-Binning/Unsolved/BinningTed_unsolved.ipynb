```python
# Figure out the minimum and maximum views for a TED Talk
min = ted_df["views"].min()
print(min)
ted_df["views"].max()
```

```python
# Slice the data and place it into bins
ted_df["View Group"] = pd.cut(ted_df["views"], bins, labels=group_labels)
ted_df.head(10)
```
